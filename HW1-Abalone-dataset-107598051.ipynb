# Step 1
### Use Pandas to load Abalone Data Set as the  dataframe from UCI Machine Learning Repository.  
**[Pandas](https://pandas.pydata.org/)**  
**[Abalone Data Set](https://archive.ics.uci.edu/ml/datasets/Abalone)**

```python
import pandas as pd
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data', 
                 header=None,names=['Sex','Length','Diameter','Height','Whole weight','Shucked weight','Viscera weight','Shell weight','Rings'])
```

# Step 6
### Plot the distribution of three types of Abalone using different colors where x-axis is the diameter and y-axis the whole weight

```python
#import matplotlib.pyplot as plt

df_M = df[(df.Sex == "M")]
df_M_W = df_M.sort_values("Whole weight")

df_F = df[(df.Sex == "F")]
df_F_W = df_F.sort_values("Whole weight")

df_I = df[(df.Sex == "I")]
df_I_W = df_I.sort_values("Whole weight")

plt.plot(df_M_W.loc[:]["Whole weight"], df_M_W.loc[:]["Diameter"],'r--',
        df_F_W.loc[:]["Whole weight"], df_F_W.loc[:]["Diameter"],'bs',
        df_I_W.loc[:]["Whole weight"], df_I_W.loc[:]["Diameter"],'g^')
plt.title('Proportion')
plt.xlabel('Diameter')
plt.ylabel('Whole weight')
plt.legend
plt.show()

plt.figure(1,figsize=(9,3))
plt.subplot(131)
plt.plot(df_M_W.loc[:]["Whole weight"], df_M_W.loc[:]["Diameter"],'r--')
plt.title('Proportion_Male')
plt.xlabel('Diameter')
plt.ylabel('Whole weight')

plt.subplot(132)
plt.plot(df_I_W.loc[:]["Whole weight"], df_I_W.loc[:]["Diameter"],'bs')
plt.title('Proportion_Female')
plt.xlabel('Diameter')
plt.ylabel('Whole weight')

plt.subplot(133)
plt.plot(df_F_W.loc[:]["Whole weight"], df_F_W.loc[:]["Diameter"],'g^')
plt.title('Proportion_Infant')
plt.xlabel('Diameter')
plt.ylabel('Whole weight')
plt.show()
```

# Step 7
### Show the proportion of three types of Abalone using pie chart.

```python
lebals = ["Male","Female","Infant"]

M = df[(df.Sex == "M")].shape[0]
F = df[(df.Sex == "F")].shape[0]
I = df[(df.Sex == "I")].shape[0]
Total = M+F+I
size = [M, F, I]

print( "M =",M,"F =",F,"I =",I ,"Total =",Total)
plt.pie(size, labels = lebals, autopct='%1.1f%%')
plt.axis("equal")
plt.show()
```
