#### Import pandas and numpy

```python
from itertools import islice
import pandas as PD
import numpy as NP
obj_Ds= PD.read_csv("ATM00756_20190514183058.csv");
obj_Ds.head()
```

#### Generate the matrixs of A and b of the matrix approach

```python
int_Limit= 798;
dbarr_A=[];
dbarr_B=[];
for int_Ct, obj_Row in enumerate(islice(obj_Ds.values, int_Limit)):
    intarr_Temp= (obj_Row[2:12].tolist());
    intarr_Temp.insert(0,1);
    dbarr_A.append(intarr_Temp);
    dbarr_B.append([obj_Row[12]]);
```

#### A

```python
print(dbarr_A);
```

#### b

```python
print(dbarr_B);
```

#### Tranform the lists to arrays

```python
objarr_A= NP.array(dbarr_A);
objarr_B= NP.array(dbarr_B);
```

#### Calculate and print the coefficents

```python
objarr_PatF= (objarr_A.T).dot(objarr_A);
objarr_PatF= NP.linalg.inv(objarr_PatF);
objarr_PatB=(objarr_A.T).dot(objarr_B);
objarr_Coef= objarr_PatF.dot(objarr_PatB);
```

### 用scikit-learn運行Ridge回歸，使用此方法在AT.A的Inverse matrix不能計算時，保證能求得迴歸引數。
#### (安裝套件:scikit-learn)

### 1.10個值中選擇一個最優的alpha值

```python
from sklearn.linear_model import RidgeCV
ridgecv = RidgeCV(alphas=[0.01, 0.1, 0.5, 1, 3, 5, 7, 10, 20, 100 ])
ridgecv.fit(objarr_A, objarr_B)
best_alpha = ridgecv.alpha_  
best_alpha
```

### 2.將alpha帶入Ridge回歸函數中運算，如此方式可得出其線性回歸的係數

```python
from sklearn.linear_model import Ridge
ridge = Ridge(alpha = best_alpha)
ridge.fit(objarr_A, objarr_B)

print(ridge.coef_)
# print(ridge.intercept_)
```
