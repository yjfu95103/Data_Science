## 前十進站的人數

```python
df_top10_in = df_array_table.sort_values('Entra',ascending= False).head(10)
df_top10_in
```

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
    
objects = df_top10_in.loc[:,"ID"]
y_pos = np.arange(len(objects))
performance = df_top10_in.loc[:,"Entra"]
 
plt.figure(figsize=(10,5))    
plt.bar(y_pos, performance, align='center', alpha=0.5 , color='b')
plt.xticks(y_pos, objects)
plt.ylabel('Passengers(100millions)')
plt.title('top10 Entrance Passengers')
 
plt.show()
```
